  <h2><b>üìÅ Dataset Overview:</b></h2>
    <p style="font-size: 19px;"><b>üìùDomain:</b> Public Safety & Transportation Analytics.</p>
    
   <p style="font-size: 17px;">This dataset comprises <b>307,973 records</b> and <b>21 columns</b>, covering road traffic accident details in the UK. It includes accident timing, severity, environmental factors, vehicle involvement, and geolocation.</p>
    <p style="font-size: 19px;"><b>üìå Data Source:</b> <a href="https://www.kaggle.com/datasets/atharvasoundankar/road-accidents-dataset" target="_blank">Road Accidents Dataset ‚Äì Kaggle</a></p>
   <p style="font-size: 19px;">üìÅ <b>Format:</b> CSV (Comma Separated Values)</p>

 


  <h3><b>üîç Key Columns or Variety:</b></h3>
  <ul style="font-size: 16px;">
    <li><b>Accident_Index</b>: Unique identifier for each accident case.</li>
    <li><b>Accident Date</b>: The date on which the accident occurred.</li>
    <li><b>Day_of_Week</b>: The day (Monday‚ÄìSunday) when the accident took place.</li>
    <li><b>Time</b>: Time of the accident (format: HH:MM).</li>
    <li><b>Accident_Severity</b>: Categorized as Fatal, Serious, or Slight.</li>
    <li><b>Number_of_Casualties</b>: Total casualties involved per accident.</li>
    <li><b>Number_of_Vehicles</b>: Number of vehicles involved.</li>
    <li><b>Vehicle_Type</b>: Type of vehicle involved (e.g., Car, Motorcycle).</li>
    <li><b>Weather_Conditions</b>: Weather at the time of the accident (e.g., Fog, Rain, Fine).</li>
    <li><b>Road_Surface_Conditions</b>: Surface state like Dry, Wet, Icy.</li>
    <li><b>Light_Conditions</b>: Conditions like Daylight or Darkness.</li>
    <li><b>Junction_Detail</b> & <b>Junction_Control</b>: Structure and control of junctions.</li>
    <li><b>Carriageway_Hazards</b>: Hazards such as obstructions or spillages.</li>
    <li><b>Police_Force</b>: The policing authority responsible for the accident record.</li>
    <li><b>Local_Authority_(District)</b>: The local district where the accident occurred.</li>
    <li><b>Road_Type</b>: Types like Single carriageway, One way street, etc.</li>
    <li><b>Speed_limit</b>: Legal speed limit at the site.</li>
    <li><b>Latitude & Longitude</b>: Geolocation of the accident.</li>
    <li><b>Urban_or_Rural_Area</b>: Indicates if the accident occurred in an urban or rural area.</li>
  </ul>

  <p style="font-size: 19px;">üßπ <b>Data Quality:</b> Contains missing values and formatting issues requiring data cleaning, making it suitable for preprocessing tasks</p>

### 2.10 Display All Unique Values Which Are Less Than 100.

<div style="border: 2px solid #007acc; padding: 1px; border-radius: 10px; background-color: #f0f8ff; text-align: left;"> </div>

```python
# Ensure all columns and rows display without truncation
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Loop through all columns and display unique values
sl_no = 0
for col in df.columns:
    unq_cnt = df[col].nunique(dropna=True)
    dtype = df[col].dtype
    sl_no += 1
    print(f"\nüîπ Column {sl_no}: {col}, Data Type - {dtype}")
    print(f"Unique Values ({unq_cnt}):")
    if unq_cnt < 100:
        print(df[col].unique())
    else:
        print('*** More number so not printing')
```
