```python
# Custom function to check for various types of "missing" data
def check_missing_blanks(df):
    summary = []

    for col in df.columns:
        total_rows = len(df)
        nan_count = df[col].isna().sum()
        none_count = (df[col] == 'None').sum()
        empty_count = (df[col] == '').sum()
        space_count = df[col].apply(lambda x: isinstance(x, str) and x.strip() == '').sum()

        total_missing = nan_count + none_count + empty_count + space_count

        if total_missing > 0:
            summary.append({
                'Column': col,
                'NaN': nan_count,
                "'None'": none_count,
                "'' (empty)": empty_count,
                "' ' or whitespace": space_count,
                'Total Missing-like': total_missing,
                'Total Rows': total_rows
            })

    return pd.DataFrame(summary)

# Call the function
missing_summary = check_missing_blanks(df)
missing_summary
```

```python
# Step: Convert time to minutes since midnight
def time_to_minutes(t):
    if pd.isnull(t):
        return np.nan
    return t.hour * 60 + t.minute

df['Time_in_Minutes'] = df['Time'].apply(time_to_minutes)
```

```python
df.head()
```

```python
# Step 4: Calculate daily average time (in minutes)
daily_mean_time = df.groupby('Accident_Date')['Time_in_Minutes'].mean()
```

```python
# Step 5: Function to fill missing time based on mean of that day
def fill_missing_time(row):
    if pd.isnull(row['Time']):  # If 'Time' is missing
        mean_minutes = daily_mean_time.get(row['Accident_Date'], np.nan)   # Get the average time for that date
        if pd.isna(mean_minutes):   # If there's no mean for the date, return NaN
            return np.nan
        # Otherwise, convert the mean minutes to HH:MM and return it
        return datetime.time(int(mean_minutes // 60), int(mean_minutes % 60))
    return row['Time']   # If time is already present, return it unchanged

# Step 6: Apply function to fill 'Time_Filled'
df['Time'] = df.apply(fill_missing_time, axis=1)
```

### 2.3.3.2 Categorical vs Continuous Column Analysis.

<div style="border: 2px solid #007acc; padding: 2px; border-radius: 10px; background-color: #f0f8ff; text-align: left; width : 43%"> </div>

##### Need to Convert Columns to category Type because of
##### 🔄 Efficient Memory Usage – Pandas stores category data more efficiently.
##### 📈 Model Performance – Many ML algorithms (like decision trees) handle categories differently.
##### 📊 Better Summary Stats – Easier grouping and aggregation in groupby() operations.
##### 🎨 Visualization – Categories make plots like bar charts and box plots easier to interpret.
##### 📌 Convert when:
##### Column is of object type and has low number of unique values.
##### The values are labels, not numbers (e.g., Weather, Road Type, Light Conditions).

##### You plan to use the column in grouping or modeling.
##### ❌ When NOT to Convert?
##### If the column has too many unique values (like IDs, timestamps, or names).
##### If the column is numeric and represents continuous data.

```python
# Prepare list to collect summary
summary = []

# Loop through each column
for col in df.columns:
    unique_vals = df[col].nunique(dropna=True)
    dtype = df[col].dtype
    
    # Suggest type
    if dtype == 'object' and unique_vals < 20:
        suggestion = 'Categorical'
    elif pd.api.types.is_numeric_dtype(df[col]):
        suggestion = 'Continue as ' + str(dtype)
    else:
        suggestion = 'Needs Review'
    
    # Append to summary list
    summary.append({
        'Column': col,
        'Data Type': str(dtype),
        'Unique Values': unique_vals,
        'Suggested Type': suggestion
    })

# Convert to DataFrame for display
summary_df = pd.DataFrame(summary)

# Display sorted by unique values
#summary_df = summary_df.sort_values(by='Unique Values')
summary_df
```

```python
# Ensure full display
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Dictionary to store sorted unique values
unique_values_dict = {}
column_headers_with_counts = []

# Loop through all columns and collect sorted unique values
for col in df.columns:
    unique_vals = df[col].dropna().unique()
    unique_count = len(unique_vals)

    # Sort and store if fewer than 500 unique values
    if unique_count < 500:
        sorted_vals = sorted(unique_vals)
        unique_values_dict[col] = pd.Series(sorted_vals)
    else:
        unique_values_dict[col] = pd.Series(['*** More than 500 unique values, skipped ***'])

    # Add unique count to column name
    column_headers_with_counts.append(f"{col} ({unique_count})")

# Create the DataFrame
unique_df = pd.DataFrame(unique_values_dict)
unique_df.columns = column_headers_with_counts

# Export to Excel
output_path = "C:/Entri/Project/data/unique_column_values.xlsx"

unique_df.to_excel(output_path, index=False)

# Display the first few rows
unique_df.head()
```

```python
# Mark multi-vehicle accidents (more than 1 vehicle involved)
df['Multi_Vehicle'] = df['Number_of_Vehicles'] > 1

# Group 
grouped = (
    df.groupby([df['Accident_DateTime'].dt.year,
                'Junction_Control',
                'Accident_Severity'], observed=False)
    .agg(
        Accident_Count=('Accident_DateTime', 'count'),
        Total_Vehicles_Sum=('Number_of_Vehicles', 'sum'),
        Total_Vehicles_Mean=('Number_of_Vehicles', 'mean'),
        Multi_Vehicle_Count=('Multi_Vehicle', 'sum')
    )
)

# Add percentage column
grouped['Multi_Vehicle_%'] = (
    grouped['Multi_Vehicle_Count'] / grouped['Accident_Count'] * 100
).round(1)

# Drop the raw count column if not needed
grouped = grouped.drop(columns='Multi_Vehicle_Count')

# Sort
grouped = grouped.sort_values(
    by=['Accident_DateTime', 'Total_Vehicles_Mean'],
    ascending=[True, False]
)

# Display result
grouped
```

## 2.3.8 Filtering.

<div style="border: 2px solid #007acc; padding: 3px; border-radius: 10px; background-color: #f0f8ff; text-align: left; width : 15%;"> </div>

### Extract the Accident Records Based on Year/Month, Day/Night, different Time Period wise Filtered.

<div style="border: 2px solid #007acc; padding: 1px; border-radius: 10px; background-color: #f0f8ff; text-align: left; width : 93%"> </div>

```python
# mode is year or month that makes the report different case of sorting only.
# it helps in case of comparison
def show_time_period_data(df, mode="year", month=None, day_night=None, time_period=None):
    temp_df = df.copy()

    # Extract datetime parts
    temp_df['Hour'] = temp_df['Accident_DateTime'].dt.hour
    temp_df['Year'] = temp_df['Accident_DateTime'].dt.year
    temp_df['Month'] = temp_df['Accident_DateTime'].dt.month_name()
    temp_df['Day_of_Week'] = temp_df['Accident_DateTime'].dt.day_name()

    # Day/Night classification
    temp_df['Day_Night'] = temp_df['Hour'].apply(
        lambda h: 'Unknown' if pd.isna(h) else 'Day' if 6 <= h < 18 else 'Night'
    )

    # Time period classification
    def time_of_day(hour):
        if pd.isna(hour): return 'Unknown'
        elif 0 <= hour < 4: return 'Mid Night'
        elif 4 <= hour < 6: return 'Early Morning'
        elif 6 <= hour < 9: return 'Morning Rush'
        elif 9 <= hour < 12: return 'Forenoon'
        elif 12 <= hour < 14: return 'Noon'
        elif 14 <= hour < 17: return 'Afternoon'
        elif 17 <= hour < 20: return 'Evening Rush'
        elif 20 <= hour < 22: return 'Late Evening'
        else: return 'Night'

    temp_df['Time_Period'] = temp_df['Hour'].apply(time_of_day)

    # Mapping & orders
    period_label_map = {
        'Mid Night': 'Mid Night (00:00 - 03:59)',
        'Early Morning': 'Early Morning (04:00 - 05:59)',
        'Morning Rush': 'Morning Rush (06:00 - 08:59)',
        'Forenoon': 'Forenoon (09:00 - 11:59)',
        'Noon': 'Noon (12:00 - 13:59)',
        'Afternoon': 'Afternoon (14:00 - 16:59)',
        'Evening Rush': 'Evening Rush (17:00 - 19:59)',
        'Late Evening': 'Late Evening (20:00 - 21:59)',
        'Night': 'Night (22:00 - 23:59)'
    }
    period_order = list(period_label_map.keys())
    weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    day_night_order = ['Day', 'Night']
    month_order = ['January','February','March','April','May','June',
                   'July','August','September','October','November','December']

    # Set categorical orders
    temp_df['Time_Period'] = pd.Categorical(temp_df['Time_Period'], categories=period_order, ordered=True)
    temp_df['Day_of_Week'] = pd.Categorical(temp_df['Day_of_Week'], categories=weekday_order, ordered=True)
    temp_df['Day_Night'] = pd.Categorical(temp_df['Day_Night'], categories=day_night_order, ordered=True)
    temp_df['Month'] = pd.Categorical(temp_df['Month'], categories=month_order, ordered=True)

    # Apply filters
    if month:
        temp_df = temp_df[temp_df['Month'] == month]
    if day_night:
        temp_df = temp_df[temp_df['Day_Night'] == day_night]
    if time_period:
        temp_df = temp_df[temp_df['Time_Period'] == time_period]

    # Time period display labels
    temp_df['Time_Period_Display'] = temp_df['Time_Period'].map(period_label_map)

    # Decide grouping keys based on mode
    if mode == "year":
        group_keys = ['Year']
    elif mode == "month":
        group_keys = ['Month', 'Year']
    else:
        raise ValueError("mode must be 'year' or 'month'")

    # Grouping
    day_night_counts = (
        temp_df.groupby(group_keys + ['Day_Night'], observed=False)
        .size()
        .unstack(fill_value=0)
        [day_night_order]
    )

    weekday_counts = (
        temp_df.groupby(group_keys + ['Day_of_Week'], observed=False)
        .size()
        .unstack(fill_value=0)
        [weekday_order]
    )

    time_period_counts = (
        temp_df.groupby(group_keys + ['Time_Period'], observed=False)
        .size()
        .unstack(fill_value=0)
        [period_order]
    )
    time_period_counts.columns = [period_label_map[p] for p in period_order]

    # Sort
    if mode == "year":
        day_night_counts = day_night_counts.sort_index(level=[0])
        weekday_counts = weekday_counts.sort_index(level=[0])
        time_period_counts = time_period_counts.sort_index(level=[0])
    else:  # month mode
        day_night_counts = day_night_counts.sort_index(level=[0,1])
        weekday_counts = weekday_counts.sort_index(level=[0,1])
        time_period_counts = time_period_counts.sort_index(level=[0,1])

    # Display
    from IPython.display import display
    display(day_night_counts)
    display(weekday_counts)
    display(time_period_counts)
```
